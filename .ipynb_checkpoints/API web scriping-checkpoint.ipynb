
Step 7 - Put everything together - Loop through results and append data inside a list

```python
name = []
image = []
priceShow = []
discount = []
ratingScore = []
review = []
location = []
description = []
sellerName = []
sellerId = []
brandName = []
inStock = []
originalPrice = []

#using loop
for result in result_json:
    # name = []
    try:
        name.append(result['name'].replace('&#8482;',''))
    except:
        name.append('')
    # image = []
    try:
        image.append(result['image'].replace('&#8482;',''))
    except:
        image.append('')
    # priceShow = []  
    try:
        priceShow.append(result['priceShow'].replace("৳", ""))
        
    except:
        priceShow.append('')
    # discount = []
    try:
        discount.append(result['discount'].replace('&#8482;',''))
    except:
        discount.append('')
    # ratingScore = []
    try:
        ratingScore.append(result['ratingScore'].replace('&#8482;',''))
    except:
        ratingScore.append('')
    # review = []
    try:
        review.append(result['review'].replace('&#8482;',''))
    except:
        review.append('')
    # location = []
    try:
        location.append(result['location'].replace('&#8482;',''))
    except:
        location.append('')
    # description = []
    try:
        description.append(result['description'])
    except:
        description.append('')
    # sellerName = []
    try:
        sellerName.append(result['sellerName'].replace('&#8482;',''))
    except:
        sellerName.append('')
    # sellerId = []
    try:
        sellerId.append(result['sellerId'].replace('&#8482;',''))
    except:
        sellerId.append('')
    # brandName = []
    try:
        brandName.append(result['brandName'])
    except:
        brandName.append('')
    # inStock = []
    try:
        inStock.append(result['inStock'])
    except:
        inStock.append('')
    # originalPrice = []
    try:
        originalPrice.append(result['originalPrice'])
    except:
        originalPrice.append('')
```

```python
description
```

Step 8 - Pandas Dataframe

```python
daraz_data = pd.DataFrame({'Product Name': name, 'Product Image': image, 'Discount Price': priceShow, 'Total Discount': discount, 'Rating': ratingScore, 'Review': review,'Store location': location, 'Description': description, 'Seller Name':sellerName, 'Seller Id': sellerId, 'Brand Name': brandName, 'Stock': inStock, 'Original Price': originalPrice})
```

```python
daraz_data
```

```python
daraz_data.to_csv('daraz_products.csv', index=False)
```
